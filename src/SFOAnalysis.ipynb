```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/leases.csv')
df
```

```python
df_sf = df[df['market'] == 'San Francisco']
df_sf
```

```python
quarter_map = {
    'Q1': '-01-01',  # January 1st
    'Q2': '-04-01',  # April 1st
    'Q3': '-07-01',  # July 1st
    'Q4': '-10-01'   # October 1st
}

df_sf_abs = df_sf[["year", "quarter", "RBA", "leasedSF"]]
df_sf_abs = df_sf_abs.groupby(['year', 'quarter']).agg({
    'RBA' : 'mean',
    'leasedSF' : 'mean'
}).reset_index()
df_sf_abs['absorption'] = df_sf_abs['leasedSF'] / df_sf_abs['RBA'] * 100
df_sf_abs['Date'] = df_sf_abs['year'].astype(str) + df_sf_abs['quarter'].map(quarter_map)
df_sf_abs['Date'] = pd.to_datetime(df_sf_abs['Date'])

df_sf_abs
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=df_sf_abs, x='Date', y='absorption')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler

# First, let's separate the data into training and testing sets
train_data = df_sf_abs[df_sf_abs['year'] < 2024].copy()
test_data = df_sf_abs[df_sf_abs['year'] == 2024].copy()

# Feature engineering - create some time-based features
# For both training and testing sets
for df in [train_data, test_data]:
    df['year_num'] = df['year'] - df['year'].min()  # Years since start
    df['quarter_num'] = df['quarter'].str[1].astype(int)  # Extract quarter number

# Define features and target
features = ['year_num', 'quarter_num', 'RBA']
target = 'absorption'

# Initialize and scale the data
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_train = scaler_X.fit_transform(train_data[features])
y_train = scaler_y.fit_transform(train_data[[target]])

X_test = scaler_X.transform(test_data[features])
y_test = test_data[target].values.reshape(-1, 1)

# Train a Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train.ravel())

# Make predictions on test data
y_pred_scaled = rf_model.predict(X_test).reshape(-1, 1)
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Test Performance Metrics:")
print(f"MSE: {mse:.6f}")
print(f"RMSE: {rmse:.6f}")
print(f"MAE: {mae:.6f}")
print(f"R²: {r2:.6f}")

# Create future dates for 2025-2026
future_quarters = []
for year in [2025, 2026]:
    for q in range(1, 5):
        future_quarters.append({
            'year': year,
            'quarter': f'Q{q}',
            'year_num': year - train_data['year'].min(),
            'quarter_num': q
        })

future_df = pd.DataFrame(future_quarters)

# Make predictions for RBA (since we need this feature for our model)
# This is a simple approach - we could make this more sophisticated
last_rba = df_sf_abs['RBA'].iloc[-1]
future_df['RBA'] = last_rba * 1.02  # Assuming 2% growth per quarter

# Make predictions for future dates
X_future = scaler_X.transform(future_df[features])
future_pred_scaled = rf_model.predict(X_future).reshape(-1, 1)
future_pred = scaler_y.inverse_transform(future_pred_scaled.reshape(-1, 1))

future_df['absorption_forecast'] = future_pred

# Create visualization of actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(train_data['Date'], train_data['absorption'], label='Training Data')
plt.plot(test_data['Date'], test_data['absorption'], label='Actual Test Data')
plt.plot(test_data['Date'], y_pred, 'r--', label='Predicted Test Data')

# Add future predictions
# We need to create dates for the future dataframe
quarter_map = {
    'Q1': '-01-01',
    'Q2': '-04-01',
    'Q3': '-07-01',
    'Q4': '-10-01'
}
future_df['Date'] = pd.to_datetime(future_df['year'].astype(str) + 
                                  future_df['quarter'].map(quarter_map))
plt.plot(future_df['Date'], future_df['absorption_forecast'], 'g--', label='Future Forecast (2025-2026)')

plt.title('Absorption: Actual vs Predicted vs Forecast')
plt.xlabel('Date')
plt.ylabel('Absorption')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Display the forecast results
print("\nForecast for 2025-2026:")
print(future_df[['year', 'quarter', 'Date', 'absorption_forecast']])

# Display test predictions alongside actual values for comparison
test_results = pd.DataFrame({
    'Date': test_data['Date'],
    'Actual': y_test.flatten(),
    'Predicted': y_pred.flatten(),
    'Error': y_test.flatten() - y_pred.flatten()
})
print("\nTest Results (2024):")
print(test_results)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')  # Suppress convergence warnings

# First, let's separate the data into training and testing sets
train_data = df_sf_abs[df_sf_abs['year'] < 2024].copy()
test_data = df_sf_abs[df_sf_abs['year'] == 2024].copy()

# For ARIMA, we'll focus on the absorption values as a time series
train_series = train_data.set_index('Date')['absorption']
test_series = test_data.set_index('Date')['absorption']

# Fit ARIMA model - we'll use (2,1,2) as parameters, but you may want to tune these
# (p=2 for autoregression, d=1 for differencing, q=2 for moving average)
model = ARIMA(train_series, order=(2, 1, 2))
model_fit = model.fit()

# Print model summary
print("ARIMA Model Summary:")
print(model_fit.summary())

# Make predictions for the test period (2024)
# We'll forecast 4 steps ahead (4 quarters of 2024)
forecast_test = model_fit.get_forecast(steps=len(test_series))
forecast_mean = forecast_test.predicted_mean
forecast_ci = forecast_test.conf_int()

# Calculate performance metrics
y_test = test_series.values
y_pred = forecast_mean.values
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
# R² can sometimes be negative for forecasting if the model performs worse than a horizontal line
r2 = r2_score(y_test, y_pred)

print(f"\nTest Performance Metrics:")
print(f"MSE: {mse:.6f}")
print(f"RMSE: {rmse:.6f}")
print(f"MAE: {mae:.6f}")
print(f"R²: {r2:.6f}")

# Create future dates for 2025-2026 (8 quarters)
last_date = test_data['Date'].max()
future_dates = pd.date_range(start=pd.Timestamp(last_date) + pd.DateOffset(months=3), 
                            periods=8, freq='3M')

# Forecast for 2025-2026
forecast_future = model_fit.get_forecast(steps=len(test_series) + len(future_dates))
future_mean = forecast_future.predicted_mean
future_ci = forecast_future.conf_int()

# Extract only the future portion (excluding 2024 which was our test set)
future_forecast = future_mean.iloc[len(test_series):]
future_ci = future_ci.iloc[len(test_series):]

# Create a dataframe for future predictions with quarters and years
future_df = pd.DataFrame({
    'Date': future_dates,
    'absorption_forecast': future_forecast.values
})
future_df['year'] = future_df['Date'].dt.year
future_df['quarter'] = 'Q' + (future_df['Date'].dt.month / 3).astype(int).astype(str)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(train_series.index, train_series.values, label='Training Data (2018-2023)')
plt.plot(test_series.index, test_series.values, label='Actual Test Data (2024)')
plt.plot(forecast_mean.index, forecast_mean.values, 'r--', label='Predicted Test Data (2024)')
plt.plot(future_forecast.index, future_forecast.values, 'g--', label='Future Forecast (2025-2026)')

# Add confidence intervals for the forecast
plt.fill_between(future_forecast.index, 
                future_ci.iloc[:, 0], future_ci.iloc[:, 1], 
                color='g', alpha=0.1, label='95% Confidence Interval')

plt.title('Absorption: Actual vs Predicted vs Forecast using ARIMA')
plt.xlabel('Date')
plt.ylabel('Absorption')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Display the forecast results
print("\nForecast for 2025-2026:")
print(future_df[['year', 'quarter', 'Date', 'absorption_forecast']])

# Display test predictions alongside actual values for comparison
test_results = pd.DataFrame({
    'Date': test_series.index,
    'Actual': test_series.values,
    'Predicted': forecast_mean.iloc[:len(test_series)].values,
    'Error': test_series.values - forecast_mean.iloc[:len(test_series)].values
})
print("\nTest Results (2024):")
print(test_results)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# Prepare training and testing data
train_data = df_sf_abs[df_sf_abs['year'] < 2024].copy()
test_data = df_sf_abs[df_sf_abs['year'] == 2024].copy()

# Set Date as index for both datasets
train_data = train_data.set_index('Date')
test_data = test_data.set_index('Date')

# Define exogenous variables (features that might influence absorption)
exog_train = train_data[['RBA']]
exog_test = test_data[['RBA']]

# Target variable
train_series = train_data['absorption']
test_series = test_data['absorption']

# Fit SARIMAX model
# - order=(1,1,1) for non-seasonal ARIMA components
# - seasonal_order=(1,1,0,4) for quarterly seasonality (s=4 for quarterly data)
model = SARIMAX(train_series, 
                exog=exog_train,
                order=(1,1,1), 
                seasonal_order=(1,1,0,4),
                enforce_stationarity=False)
model_fit = model.fit(disp=False)

# Print summary
print("SARIMAX Model Summary:")
print(model_fit.summary())

# Predict test period (2024)
predictions = model_fit.forecast(steps=len(test_series), exog=exog_test)

# Calculate performance metrics
mse = mean_squared_error(test_series, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(test_series, predictions)
r2 = r2_score(test_series, predictions)

print(f"\nTest Performance Metrics:")
print(f"MSE: {mse:.6f}")
print(f"RMSE: {rmse:.6f}")
print(f"MAE: {mae:.6f}")
print(f"R²: {r2:.6f}")

# Create forecasts for 2025-2026
# First, we need to estimate future RBA values
last_rba = exog_test['RBA'].iloc[-1]
# Assume 2% growth per quarter in RBA (you may want to adjust this)
future_rba = [last_rba * (1.02)**(i+1) for i in range(8)]  # 8 quarters for 2025-2026

# Create a DataFrame with future dates and RBA values
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
future_dates = []
future_years = []
future_quarters = []

for year in [2025, 2026]:
    for q in range(4):
        future_years.append(year)
        future_quarters.append(quarters[q])
        month = q*3 + 1  # Convert quarter to month (Q1->1, Q2->4, Q3->7, Q4->10)
        future_dates.append(pd.Timestamp(f"{year}-{month:02d}-01"))

future_exog = pd.DataFrame({
    'RBA': future_rba
}, index=future_dates)

# Generate forecasts for future periods
future_predictions = model_fit.forecast(steps=len(future_exog), exog=future_exog)

# Create a complete DataFrame with all results
all_dates = list(train_series.index) + list(test_series.index) + list(future_exog.index)
all_actual = list(train_series) + list(test_series) + [None] * len(future_exog)
all_predicted = [None] * len(train_series) + list(predictions) + list(future_predictions)

results_df = pd.DataFrame({
    'Date': all_dates,
    'Actual': all_actual,
    'Predicted': all_predicted
})

# Plot results
plt.figure(figsize=(14, 7))
plt.plot(train_series.index, train_series, label='Training Data (2018-2023)')
plt.plot(test_series.index, test_series, label='Actual Test Data (2024)')
plt.plot(test_series.index, predictions, 'r--', label='Predicted Test Data (2024)')
plt.plot(future_exog.index, future_predictions, 'g--', label='Future Forecast (2025-2026)')

plt.title('Absorption: Actual vs Predicted vs Forecast using SARIMAX')
plt.xlabel('Date')
plt.ylabel('Absorption')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Create future results DataFrame
future_results = pd.DataFrame({
    'Date': future_dates,
    'Year': future_years,
    'Quarter': future_quarters,
    'Forecast': future_predictions
})

print("\nForecast for 2025-2026:")
print(future_results)

# Display test results
test_comparison = pd.DataFrame({
    'Date': test_series.index,
    'Actual': test_series.values,
    'Predicted': predictions.values,
    'Error': test_series.values - predictions.values,
    'Percent_Error': (test_series.values - predictions.values) / test_series.values * 100
})

print("\nTest Results (2024):")
print(test_comparison)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# Prepare data for Prophet (requires specific column names)
prophet_train = train_data.reset_index()[['Date', 'absorption']].rename(
    columns={'Date': 'ds', 'absorption': 'y'})

# Add regressors
prophet_train['rba'] = train_data.reset_index()['RBA']
prophet_train['leased'] = train_data.reset_index()['leasedSF']

# Prepare test data
prophet_test = test_data.reset_index()[['Date', 'absorption']].rename(
    columns={'Date': 'ds', 'absorption': 'y'})
prophet_test['rba'] = test_data.reset_index()['RBA']
prophet_test['leased'] = test_data.reset_index()['leasedSF']

# Initialize and train Prophet model with regressors
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    seasonality_mode='multiplicative'
)
model.add_regressor('rba')
model.add_regressor('leased')
model.fit(prophet_train)

# Make predictions on test data
forecast = model.predict(prophet_test)

# Calculate performance metrics
y_test = prophet_test['y'].values
y_pred = forecast['yhat'].values
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Prophet Test Performance Metrics:")
print(f"MSE: {mse:.6f}")
print(f"RMSE: {rmse:.6f}")
print(f"MAE: {mae:.6f}")
print(f"R²: {r2:.6f}")

# Create future dataframe for 2025-2026 predictions
future_dates = []
future_years = []
future_quarters = []
future_rba = []
future_leased = []

# Get last values to project forward
last_rba = test_data['RBA'].iloc[-1]
last_leased = test_data['leasedSF'].iloc[-1]

for i, year in enumerate([2025, 2026]):
    for q in range(1, 5):
        future_years.append(year)
        future_quarters.append(f"Q{q}")
        month = (q - 1) * 3 + 1
        future_dates.append(pd.Timestamp(f"{year}-{month:02d}-01"))
        
        # Project RBA and leasedSF with modest growth
        quarters_ahead = i * 4 + q
        future_rba.append(last_rba * (1.015) ** quarters_ahead)
        future_leased.append(last_leased * (1.01) ** quarters_ahead)

# Create future dataframe
future_df = pd.DataFrame({
    'ds': future_dates,
    'rba': future_rba,
    'leased': future_leased,
    'year': future_years,
    'quarter': future_quarters
})

# Generate forecasts
future_fcst = model.predict(future_df)

# Plot results
plt.figure(figsize=(14, 7))
plt.plot(prophet_train['ds'], prophet_train['y'], 'b-', label='Training Data (2018-2023)')
plt.plot(prophet_test['ds'], prophet_test['y'], 'k-', label='Actual Test Data (2024)')
plt.plot(prophet_test['ds'], forecast['yhat'], 'r--', label='Predicted Test Data (2024)')
plt.plot(future_df['ds'], future_fcst['yhat'], 'g--', label='Future Forecast (2025-2026)')

# Add confidence intervals for future forecast
plt.fill_between(future_fcst['ds'], 
                future_fcst['yhat_lower'], 
                future_fcst['yhat_upper'], 
                color='g', alpha=0.1, label='95% Confidence Interval')

plt.title('Absorption: Actual vs Predicted vs Forecast using Prophet with Regressors')
plt.xlabel('Date')
plt.ylabel('Absorption x 100')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Add annotations for future predictions
for i, date in enumerate(future_df['ds']):
    plt.annotate(f"{future_fcst['yhat'].iloc[i]:.4f}", 
                 (date, future_fcst['yhat'].iloc[i]),
                 textcoords="offset points",
                 xytext=(0,10), 
                 ha='center', fontsize=6)

plt.show()

# Display forecast for 2025-2026
future_results = pd.DataFrame({
    'Date': future_dates,
    'Year': future_years,
    'Quarter': future_quarters,
    'Forecast': future_fcst['yhat'].values,
    'Lower_95': future_fcst['yhat_lower'].values,
    'Upper_95': future_fcst['yhat_upper'].values
})

print("\nForecast for 2025-2026:")
print(future_results[['Date', 'Year', 'Quarter', 'Forecast', 'Lower_95', 'Upper_95']])

# Display test results
test_comparison = pd.DataFrame({
    'Date': prophet_test['ds'],
    'Actual': prophet_test['y'],
    'Predicted': forecast['yhat'],
    'Error': prophet_test['y'] - forecast['yhat'],
    'Percent_Error': (prophet_test['y'] - forecast['yhat']) / prophet_test['y'] * 100
})

print("\nTest Results (2024):")
print(test_comparison)

# Plot components to understand drivers
plt.figure(figsize=(12, 10))
model.plot_components(forecast)
plt.tight_layout()
plt.show()
```
