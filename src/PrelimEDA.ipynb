```python
quarter_map = {
    'Q1': '-01-01',  # January 1st
    'Q2': '-04-01',  # April 1st
    'Q3': '-07-01',  # July 1st
    'Q4': '-10-01'   # October 1st
}
df['Date'] = df['year'].astype(str) + df['quarter'].map(quarter_map)
df['Date'] = pd.to_datetime(df['Date'])
df_overall_avg = df.groupby(['year', 'quarter', 'region']).agg({'overall_rent': 'mean'}).reset_index()
df_overall_avg['Date'] = df_overall_avg['year'].astype(str) + df_overall_avg['quarter'].map(quarter_map)
df_overall_avg['Date'] = pd.to_datetime(df_overall_avg['Date'])
df_overall_avg
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_overall_avg['Date'], df_overall_avg['overall_rent'], marker='o', linestyle='-', color='b')
plt.title('Average Rent per Quarter')
plt.xlabel('Date')
plt.ylabel('Average Value')
plt.grid(True)

# Display the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define a fixed inflation rate (e.g., 3.2% annually)
inflation_rate = 0.032  # Annual inflation

# Create a figure
plt.figure(figsize=(12, 8))

# Loop through each region to plot rent data and inflation line
for region in df_overall_avg['region'].unique():
    region_data = df_overall_avg[df_overall_avg['region'] == region]
    
    # Plot the actual rent data for the region
    plt.plot(region_data['Date'], region_data['overall_rent'], marker='o', linestyle='-', label=region)
    
    # Get the starting rent value for Q1 2018
    start_rent = region_data[region_data['Date'].dt.year == 2018]['overall_rent'].iloc[0]
    
    # Generate quarterly dates
    inflation_dates = pd.date_range(start='2017-12-31', end='2024-12-31', freq='Q')  # Quarterly

    # Adjust inflation for quarterly growth instead of annual
    quarters_since_2018 = [(date.year - 2018) * 4 + (date.quarter - 1) for date in inflation_dates]
    inflation_line = [start_rent * ((1 + inflation_rate) ** (q / 4)) for q in quarters_since_2018]  # Adjusted per quarter

    # Plot the inflation trend
    plt.plot(inflation_dates, inflation_line, linestyle='--', label=f'{region} Inflation', color='black')

plt.title('Average Rent per Quarter by Region')
plt.xlabel('Date')
plt.ylabel('Average Rent')
plt.grid(True)
plt.legend(title='Region')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import MultipleLocator


# Define annual inflation rates
inflation_rates = {
    2018: 0.019,  # 1.9%
    2019: 0.023,  # 2.3%
    2020: 0.014,  # 1.4%
    2021: 0.070,  # 7.0%
    2022: 0.065,  # 6.5%
    2023: 0.034,  # 3.4%
    2024: 0.029   # 2.9%
}

# Create a figure
plt.figure(figsize=(12, 8))

# Loop through each region to plot rent data and inflation-adjusted line
for region in df_overall_avg['region'].unique():
    region_data = df_overall_avg[df_overall_avg['region'] == region]
    
    # Plot the actual rent data for the region
    plt.plot(region_data['Date'], region_data['overall_rent'], marker='o', linestyle='-', label=region)

      # Get the starting rent value for Q1 2018
    start_rent = region_data[region_data['Date'].dt.year == 2018]['overall_rent'].iloc[0]
    
    # Generate quarterly dates
    inflation_dates = pd.date_range(start='2017-12-31', end='2024-12-31', freq='Q')  # Quarterly

    # Compute inflation-adjusted rent over time using varying rates
    inflation_line = []
    prev_rent = start_rent  # Start with the first value

    for date in inflation_dates:
        year = date.year
        quarter = date.quarter
        
        # Use the correct inflation rate for the given year
        annual_rate = inflation_rates.get(year, 0.02)  # Default to 2% if year is missing
        quarterly_rate = (1 + annual_rate) ** (1/4)  # Convert annual rate to quarterly
        
        # Apply inflation rate to get new rent value
        new_rent = prev_rent * quarterly_rate
        inflation_line.append(new_rent)
        
        # Update for next quarter
        prev_rent = new_rent

    # Plot the inflation trend
    plt.plot(inflation_dates, inflation_line, linestyle='--', label=f'{region} Inflation', color='black')
    
plt.title('Average Rent per Quarter by Region')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Average Overall Rent ($)', fontsize=14,labelpad=15)
plt.grid(True)
plt.axvline(pd.Timestamp('2020-04-01'), color='red', linestyle='--', label='COVID')
plt.legend(title='Legend')
handles, labels = plt.gca().get_legend_handles_labels()

# Keep only the unique region names, or filter out unwanted labels like 'COVID Inflation'
filtered = [(h, l) for h, l in zip(handles, labels) if "Inflation" not in l]
# Unzip the filtered pairs
handles, labels = zip(*filtered)
plt.legend(handles, labels, title="Region", fontsize=12, title_fontsize=13, loc='upper left')


ax = plt.gca()  # Get current axis
ax.yaxis.set_major_locator(MultipleLocator(2))  # Adds a tick every 50 units


plt.tight_layout()
plt.show()
```
